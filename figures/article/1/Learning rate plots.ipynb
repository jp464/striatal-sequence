```python
import openpyxl
import numpy as np
import matplotlib.pyplot as plt 
from matplotlib.pyplot import figure
plt.rcParams["figure.figsize"] = (13, 18)
```

```python
filename = "/Users/stan.park712/MyDrive/striatum/hebbian_sequence_learning/figures/output/learning_rate_data.xlsx"
wb = openpyxl.load_workbook(filename)
sheet = wb.active
```

### 3. $A^{cc} = A^{bb} = 3.4$, changing $A^{bc} = A^{cb}$

```python
A = np.arange(1, 5.2, 0.2)
corr_ctx = [v.value for v in sheet["E"][34:55]]
spd_ctx = [v.value for v in sheet["F"][34:55]]
corr_bg = [v.value for v in sheet["G"][34:55]]
spd_bg = [v.value for v in sheet["H"][34:55]]

fig, axes = plt.subplots(2)
axes[0].plot(A, corr_ctx, label="cortex")
axes[0].plot(A, corr_bg, label="BG")
axes[0].set_xlabel(r'$A^{bc} = A^{cb}$')
axes[0].set_ylabel('Average peak correlation')
axes[0].legend()

axes[1].plot(A, spd_ctx, label="cortex")
axes[1].plot(A, spd_bg, label="BG")
axes[1].set_xlabel(r'$A^{bc} = A^{cb}$')
axes[1].set_ylabel('Average peak-peak interval time')
axes[1].legend()
plt.show()
```
